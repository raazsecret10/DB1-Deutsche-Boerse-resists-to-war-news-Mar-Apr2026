```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# the dates and their respective prices [^GDAXI] and [DB1]
data = {
    'Date': ['2026-02-28', '2026-03-02', '2026-04-07', '2026-04-08'],
    'Scenario': ['Normal', 'War Start', 'Normal', 'Ceasefire'],
    'DAX_Price': [25284.26 , 24638.00 , 22921.59 , 24080.63],  
    'DB1_Price': [232.40, 238.20, 253.70, 254.20]  
}
```

```python
df = pd.DataFrame(data)

# 2. Derive Daily Returns (%) through code
# Formula: ((New Price / Old Price) - 1) * 100
df['DAX_Return_%'] = df['DAX_Price'].pct_change() * 100
df['DB1_Return_%'] = df['DB1_Price'].pct_change() * 100
```

```python
# 3. Filter for the specific 'Day After' Events
war_day = df.iloc[1]
peace_day = df.iloc[3]
```

```python
print("* IRAN US ISRAEL 2026 WAR *")
print(f"\nWAR SHOCK date (March 2, 2026):")
print(f"  DAX Change: {war_day['DAX_Return_%']:.2f}%")
print(f"  DB1 Change: {war_day['DB1_Return_%']:.2f}%")

print(f"\nCEASEFIRE date (April 8, 2026):")
print(f"  DAX Change: {peace_day['DAX_Return_%']:.2f}%")
print(f"  DB1 Change: {peace_day['DB1_Return_%']:.2f}%")
```

```python
# 5. Simple Plot to Visualize the 'Inverse' behavior
df_plot = df.dropna()
df_plot.plot(kind='bar', x='Scenario', y=['DAX_Return_%', 'DB1_Return_%'], 
             color=["#E0600B", "#459900"], figsize=(10,6))
plt.axhline(0, color='black', linewidth=1.0)
plt.title("DB1 vs DAX: IRAN US ISRAEL 2026 WAR")
plt.ylabel("Daily Change (%)")
plt.show()
```
